```python
%reload_ext autoreload
%autoreload 2

import os
import torch
import numpy as np
from datasets import Dataset, Facebook100, FilterTopClass
from torch_geometric.utils import degree, is_undirected
from tqdm.auto import tqdm
import pandas as pd
pd.options.display.float_format = "{:,.2f}".format
pd.set_option('max_rows', None)
```

```python
target = 'status'
rows = []
for dataset in tqdm(Facebook100.available_datasets):
    data = Facebook100(root='datasets/fb100/'+dataset, name=dataset, target=target)[0]
    stat = {
        'name': dataset,
        'nodes': data.num_nodes,
        'edges': data.num_edges,
        'features': data.num_features,
        'classes': int(data.y.max().item() + 1),
        'mean degree': degree(data.edge_index[1], num_nodes=data.num_nodes).mean().item(),
        'median degree': degree(data.edge_index[1], num_nodes=data.num_nodes).median().item(),
        'directed': not is_undirected(data.edge_index, num_nodes=data.num_nodes),
        'baseline': data.y.unique(return_counts=True)[1].max().item() / data.num_nodes
    }
    rows.append(stat)
```

```python
df = pd.DataFrame(rows).set_index('name').sort_values('nodes', ascending=False)
df[df['baseline'] <= 0.7]
```
